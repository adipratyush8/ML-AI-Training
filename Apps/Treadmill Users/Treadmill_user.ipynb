# UniVariate Analysis

1. Categorical Data

```python
#Using Countpot for Categorical Data
for col in cat_col:
    print(col)
    plt.subplot(1,2,1)
    sns.countplot(data=df,x=col,palette=choice(list))
    plt.title(col)
    plt.subplot(1,2,2)
    df[col].value_counts().plot.pie(autopct='%1.1f%%',figsize=(10,4))#Creating a Pie Chart of the following Categorical Data
    plt.show()
```

```python
#Using dist,box,violin plot for better understanding
for col in num_cols:
    colour=[random(),random(),random()]
    fig=plt.figure(figsize=(10,5))
    fig.suptitle(col)
    plt.subplot(1,3,1)
    sns.distplot(df[col],color=colour)#RGB
    plt.subplot(1,3,2)
    sns.boxplot(data=df,x=col,color=colour)
    plt.subplot(1,3,3)
    sns.violinplot(data=df,x=col,color=colour)
    plt.show()
```

# Firs We will be comparing Categories with numbers using first 3 Plots


```python
for col in cat_col:
    for ncol in num_cols:
        plt.figure(figsize=(10,5))
        plt.suptitle(f'{col} vs {ncol}')
        plt.subplot(1,3,1)
        sns.boxplot(data=df,x=col,y=ncol,palette=choice(list))
        plt.subplot(1,3,2)
        sns.barplot(data=df,x=col,y=ncol,palette=choice(list))
        plt.subplot(1,3,3)
        sns.violinplot(data=df,x=col,y=ncol,palette=choice(list))
        plt.show()
    
```

```python
pg=sns.PairGrid(data=df[num_cols],palette=choice(list))
pg.map_diag(sns.histplot)
pg.map_upper(sns.kdeplot)
pg.map_lower(sns.regplot)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=df[num_cols].corr(),annot=True,cmap=choice(list))
plt.show()

```

# MultiVariate Analysis

```python
# Here We will be again using Box,bar,Violin (with hue=product) in order to make it multivariate
```

```python
num_cols
```

```python
cat_col
```

```python
for col in cat_col:
    if col == 'Product':
        continue
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col} vs {ncol}')
        plt.subplot(1,3,1)
        sns.boxplot(
            data = df, x = col, y = ncol, 
            hue = 'Product', palette=choice(list))
        plt.subplot(1,3,2)
        sns.violinplot(
            data=df, x = col, y = ncol, 
            hue = 'Product', palette=choice(list))
        plt.subplot(1,3,3)
        sns.barplot(
            data = df, x = col, y = ncol, 
            hue = 'Product', palette=choice(list))
        plt.show()
    
```
