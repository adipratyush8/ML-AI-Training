```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#This is the way to read a particular file
movie=pd.read_csv(r'mymovies.csv')
#Another way
#movie=pd.read_clipboard('')------For this you have to copy all the selected text in the file in order to make it read
```

```python
movie
```

```python
#Q1. Find the movie with the longest duration
movie['Name'][movie['Duration'].idxmax()]
#movie(needed element then condition)
```

```python
#Q2. Find the number of movies who got 10 rating
# movie['Name'][movie['Rating']==10] # This will tell you the names of the movie
sum(movie['Rating']==10)
```

```python
# Q3.How many Industries are there
# movie['Industries'].nunique()
#And in order to know names of them 
movie['Industries'].unique()
```

```python
#Q4.How many movies are there in each Industries
[movie['Industries'].value_counts()]
```

```python
#Q4. What is the least liked comedy film
movie['Name'][movie[movie["Genre"]=='comedy']['Rating'].idxmin()]
```

```python
#Q5. Find the longest duration english movie
movie['Name'][movie[movie['Language']=='English']['Duration'].idxmax()]
```

```python
#Q6. What is the Genre of Intersteller
movie[movie['Name'].str.lower()=="Intersteller".lower()]['Genre'].values[0]
```

```python
#Q7. what is the average Duration of each genre
for Genre in movie['Genre'].unique():
    print(Genre,movie[movie['Genre']==Genre]['Duration'].mean())
```

```python
#Q8. What is the average rating of each Industries
for Industries in movie["Industries"].unique():
    print(Industries,movie[movie["Industries"]==Industries]['Duration'].mean())
```

```python
#Q9. Find the average rating of Romantic movies industrie wise
for Industries in movie["Industries"].unique():
    print(Industries,movie[movie['Genre']=='Romantic']['Duration'].mean())
```

```python
#Q10. List down all the English movie with their rating in descending order
# for name in movie['Name']:
movie[movie["Language"]=="English"][["Name","Rating"]].sort_values(by="Rating",ascending=False)
```

```python
#The describe() method returns description of the data in the DataFrame.
movie.describe()
#count - The number of not-empty values.
# mean - The average (mean) value.
# std - The standard deviation.
# min - the minimum value.
# 25% - The 25% percentile*.
# 50% - The 50% percentile*.
# 75% - The 75% percentile*.
# max - the maximum value.
```

* Standard deviation is a measure that is used to quantify the amount of variation or dispersion of a set of data values.
*  Group by is used for grouping the data according to the categories and apply a function to the categories. It also helps to aggregate data efficiently.

```python
# In order to find the average duration of each Genre
print(movie.groupby('Genre')['Duration'].mean())
```

```python
# In order to find the average duration of each Genre According to their Industries
print(movie.groupby(['Genre','Industries'])['Duration'].mean())
```
