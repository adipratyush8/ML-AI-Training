# Graph function :
1. plot                             - simple line plot 
2. plot.area                        - filled version of line plot 
3. plot.bar / plot.barh             - bar graph 
4. plot.box                         - box plot 
5. plot.density                     - density line (smooth curved)
6. plot.hexbin                      - hexagon version of scatter with density
7. plot.hist                        - histogram
8. plot.kde                         - kernel density estimator (Similar density graph)                       
9. plot.pie                         - pie chart of categorical 
10. plot.scatter                    - scatter graph

```python
fig=plt.figure(figsize=(12,6))
df2.plot.area();plt.title("Area Plot")
df2.plot.bar();plt.title("Bar Plot")
df2.plot.barh();plt.title("Barh Plot")
df2.plot.box();plt.title("Box Plot")
df2.plot.density();plt.title("Density Plot")
df2.plot.hexbin(x='a',y='b',C='c',gridsize=6);plt.title("Hexbin Plot")
df2.plot.kde();plt.title("Kde Plot")# Kernel density plot
df2.plot.hist();plt.title("Hist Plot")
df2.plot.scatter(x='a',y='b');plt.title("Scatter Plot")
plt.show()
```

# Now we will be Visualizing data with the help of Graphs


# 1st Categorical Plots
* Plots are basically used for visualizing the relationship between variables. Those variables can be either be completely numerical or a category like a group, class or division.

```python
sns.countplot(data=tips,x='day')
plt.title('Count plot')
```

```python
# A countplot basically counts the categories and returns a count of their occurrences. It is one of the most simple plots provided by the seaborn library.
sns.countplot(data=tips,x='day',hue='sex')#hue is used to separate the data further using the sex category
#Count plot Gives us the count of input data column (takes a single Parameter)
plt.title('Bar plot')
#Here hue is used for comparision 
```

```python
# To use this plot we choose a categorical column for the x axis and a numerical column for the y axis and we see that it creates a plot taking a mean per categorical column. 
sns.barplot(data=tips,x='day',y='total_bill')#One Numerical and One Categorical
plt.title('Bar plot')
#By default Bargraph gives mean value
```

```python
sns.barplot(data=tips,x='day',y='total_bill',estimator=sum)
```

```python
sns.barplot(data=tips,x='day',y='total_bill',hue='time',estimator=np.mean)#np.mean is bydefault graphical representation of bargraph
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='smoker')
plt.title('Box plot')
```

```python
sns.swarmplot(data=tips,x='day',y='total_bill',hue='smoker')
plt.title('Swarm plot')
#Here No overlapping takes place btw points
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='smoker')
```

# 2nd Distriution Plots
# It provides a high-level interface for drawing attractive and informative statistical graphics

```python
sns.distplot(tips['total_bill'])
# The distplot represents the univariate distribution of data
```

```python
sns.displot(data=tips,x='total_bill',y='tip')
# univariate or bivariate distribution of data
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',hue='smoker')
#It is used to draw a plot of two variables with bivariate and univariate graphs. It basically combines two different plots.
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip');plt.show()
sns.jointplot(data=tips,x='total_bill',y='tip',hue='smoker');plt.show()
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hex');plt.show()
# kind = ‘hex’ provides the hexagonal plot and kind = 'scatter'provides the scatter on the graph.
sns.jointplot(data=tips,x='total_bill',y='tip',kind='scatter');plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.kdeplot(data=tips,x='total_bill')# Both the rug and kde plot are giving us the mean of the total bill as in this case
sns.rugplot(data=tips , x='total_bill')# Both the rug and kde plot are giving us the mean of the total bill as in this case
# A rugplot is a graph that places a dash horizontally with each occurrence of an item in a dataset.
# Areas where there is great occurrence of an item see a greater density of these dashes.
# Areas where there is little occurrence of an item see just occasional dashes.
plt.subplot(1,2,2)
sns.kdeplot(data=tips , x='total_bill',y='tip')
sns.rugplot(data=tips , x='total_bill',y='tip')
```

```python
sns.regplot(data=tips,x='total_bill',y='tip');plt.title("Regression Plot");plt.show()
# lmplot() can be understood as a function that basically creates a linear model plot
sns.lmplot(data=tips,x='total_bill',y='tip');plt.title("Linear Model Plot");plt.show()
```

# Grid Plots
1. Grids are general types of plots that allow you to map plot types to grid rows and columns,

```python
from matplotlib.pyplot import scatter
pg=sns.PairGrid(data=iris)
pg.map_diag(sns.kdeplot)
pg.map_lower(sns.scatterplot)
pg.map_upper(sns.regplot)
```

```python
pg = sns.PairGrid(data=iris)
pg.map_diag(sns.kdeplot)
pg.map_lower(sns.scatterplot)
pg.map_upper(sns.kdeplot)
```

```python
# JointGrid is the general version for the grid type of jointplot(). Jointplot by Seaborn shows a relationship in the margins between 2 variables (bivariate) and 1D profiles (univariate). 
jg=sns.JointGrid(data=tips,x='total_bill',y='tip')
jg.plot(sns.regplot,sns.distplot)
```

```python
#A FacetGrid is a multi-axes grid which allows you to plot multiple grids(rows and columns)
fg=sns.FacetGrid(data=tips , row='smoker', col='day',hue='sex')
fg.map(sns.scatterplot,'total_bill','tip').add_legend
```
