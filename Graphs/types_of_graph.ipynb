# Types of graphs
1. plot            -:It is a line graph btw two Parameters
2. Bar             -:Gives us Bar Graph of Two Parameters(optional_categorical_data)
3. Hist            -:It  gives us the frequency of one or two Parameters
4. Scatter         -:Dot Graph btw Two Parameters
5. Boxplot         -:statistical overview in form of box for a single parameter , mean , outliers , quantiles(25% ,75%) , IQR (optional - catagorical data) 
6. Stem            -:more like bar graph, lines instead of bars
7. Violinplot      -:Violin shaped statistical overview graph spread of data (more like hist) single parameter (optional cat-data)

* Categorical data is a kind of data that is utilized to gather data with comparative qualities while Numerical Data is a sort of data that communicates data as numbers. It consolidates numeric qualities to portray applicable data while downright information utilizes a distinct way to deal with express data.
* Categorical data is additionally called qualitative data while numerical data is likewise called quantitative data. This is on the grounds that categorical data is utilized to qualify data prior to ordering them as per their similarities.

# Box plot,Hist Plot And Violin plot (Single Parameter)

```python
x=np.linspace(0,5,11)#Arguments(starting ponint,ending point,number of items)
x1=np.arange(0,13,3)#Arguments(starting,ending,steps)
```

```python
x1
```

```python
x
```

```python
y=x**3
```

```python
plt.plot(x,y)
plt.title('Plot Graph') #This is used to give title to the graph
plt.show() # This is responsible for showing the graph
#This plot gives us a line graph btw two Parameters
```

```python
plt.bar(x,y)
plt.title("Bar Graph")
plt.show()
```

```python
plt.hist(x,y)# It can also work for Single parameter
plt.title("Hist Graph")
plt.show()
```

```python
plt.scatter(x,y)
plt.title("Scatter Graph")
plt.grid()# It is used to set visibility of grids inside the graph 
plt.xlabel('X---->')#Used in mentioning the axis
plt.ylabel('Y---->')#Used in mentioning the axis
plt.xticks(x)
plt.yticks(range(0,200,10))#range-Arguments(Starting , ending , steps)
plt.show()
```

```python
plt.stem(x,y)#More like Bar Graph
plt.show()
```

```python
plt.violinplot(x)
plt.title("Violin Graph")
plt.show()
```

```python
plt.boxplot(x)
plt.title("Boxplot")
plt.show()
```

# Taking another y as x**4

```python
y2=x**4
```

```python
# Both the Graphs are like this
plt.plot(x,y)
plt.show()
plt.plot(x,y2)
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
plt.plot(x,y,lw=5,marker='o',markersize=15,markerfacecolor='red')
plt.plot(x,y2,lw=3,marker='v',markersize=10,markerfacecolor='blue')
plt.title("Figure")
plt.legend(['y=x**3','y2=x**4']) #With the help of legend we can mark both the lines
plt.grid()
plt.show()
```
